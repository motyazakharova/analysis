```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
df.head()
```

```python
sns.set(style='darkgrid')
```

```python
summer_months = df[df['month'].isin([5, 6, 7, 8])].copy()
```

```python
yearly_summer_avg = summer_months.groupby('year')['temp_max'].mean().reset_index()
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(data=yearly_summer_avg, x='year', y='temp_max', 
             marker='o', color='red', linewidth=2.5)

plt.title('Average Max Temperature: May - August (2011-2020)', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Temperature (°C)', fontsize=12)
plt.xticks(yearly_summer_avg['year'])
plt.grid(True, linestyle='--', alpha=0.6)

plt.show()
```

```python
dry_days = summer_months[summer_months['precipitation'] == 'без осадков']
```

```python
dry_days_count = dry_days.groupby('year').size().reset_index(name='dry_days_count')
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(
    data=dry_days_count, 
    x='year', 
    y='dry_days_count', 
    palette='flare', 
    hue='year', 
    legend=False)

plt.title('Number of Dry Days: May - August (2011-2020)', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Count of Dry Days', fontsize=12)

plt.show()
```

```python
hot_days = summer_months[summer_months['temp_max'] >= 30]

hot_days_count = hot_days.groupby('year').size().reset_index(name='hot_day_count')


plt.figure(figsize=(10, 6))
barplot = sns.barplot(
    data=hot_days_count, 
    x='year', 
    y='hot_day_count', 
    hue='year',      
    palette='Reds',
    legend= False
)

plt.title('Number of Days with Temp >= 30°C (May–August)', fontsize=14)
plt.ylabel('Count of Hot Days')
plt.show()
```

```python
df['temp_range'] = df['temp_max'] - df['temp_min']
biggest_swing = df.loc[df['temp_range'].idxmax()]
temperature = biggest_swing['temp_range']
swing_year = biggest_swing['year']
swing_month = biggest_swing['month']
swing_day = biggest_swing['day']
print(f"The biggest swing in temperature was {temperature} degrees Celcius in {swing_day}/{swing_month}/{swing_year}")
```
